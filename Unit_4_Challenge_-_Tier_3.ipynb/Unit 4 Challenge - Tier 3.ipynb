```python
# create function that takes a dataframe of prices and a borough as an argument and returns the ratio of the average price in 1998 to average price in 2018.
# I will create this function under the assumption that the dataframe, borough_prices, has been created already.

def average_ratio(df, borough):
    
    avg_price_by_year = df.groupby(['Region','Year']).agg(np.mean)
    
    avg_1998 = avg_price_by_year.loc[(str(borough),1998),'Average House Price']
    avg_2018 = avg_price_by_year.loc[(str(borough),2018),'Average House Price']
    
    ratio = avg_1998 / avg_2018
    
    return ratio
    
```

```python
# use the function to iterate through the dataframe, borough_prices, and calculate the ratio of average house prices in 1998 and 2018.
# loop will then create a dictionary with each borough and their respective ratios.

borough_list = list(set(borough_prices['Region'].tolist()))


ratio_dict = {}
for borough in borough_list:
    ratio_dict[borough] = average_ratio(borough_prices,borough)

print(ratio_dict)
```

```python
# plot the dictionary to showcase the ratios of each borough

boroughs = list(ratio_dict.keys())
ratios = list(ratio_dict.values())

plt.xticks(rotation=90)
plt.bar(boroughs,ratios)
plt.title('1998/2018 Average House Price Ratio')
plt.show()
```
